```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 数据概况的观察
    - 使用pd.read_csv()完成数据的提取
    - DataFrame.describe()完成数据概况的观察
## 数据列的提取
    - 把pd.read_csv()提取出来的DataFrame对象保存

```python
path1 = "./data/production.csv"
path2 = "./data/production_missing_category.csv"
path3 = "./data/production_missing_num.csv"
path4 = "./data/production_missing_num_4_redshift.csv"

production = pd.read_csv(path1)
production_missing_category = pd.read_csv(path2)
production_missing_num = pd.read_csv(path3)
production_missing_num_4_redshift = pd.read_csv(path4)

print('表production:')
print(production)
print(production.describe())
print('\n\n\n')

print('表production_missing_category:')
print(production_missing_category)
print(production_missing_category.describe())
print('\n\n\n')

print('表production_missing_num:')
print(production_missing_num)
print(production_missing_num.describe())
print('\n\n\n')

print('表production_missing_num_4_redshift:')
print(production_missing_num_4_redshift)
print(production_missing_num_4_redshift.describe())
print('\n\n\n')
```

## 异常值的识别和剔除
    - 使用Z-score方法剔除异常值
## 规整化
    - 属性字段数值化,将man处理为1,woman处理为0,并且把bool值处理为0/1值
    - 归一化,把数据都归一化到[0,1]区间中,min值对应0,max值对应1
## 缺失值的填充
    - 使用聚合值填充,这里使用中位数
    
**注意:一定要在剔除异常值并且规整化之后后再填充缺失值**

```python
# 对表production使用Z-Score方法
zscore = (production - production.mean()) / production.std()
zscore_r = zscore[abs(zscore) < 3]

# 对表production进行归一化处理
new_production = zscore_r / (zscore_r.max() - zscore_r.min())
print('归一化后的new_production:')
new_production['fault_flg'] = production['fault_flg']
new_production['type'] = production['type']
print(new_production)
```

```python
# 对表production_missing_category使用Z-Score方法
zscore = (production_missing_category - production_missing_category.mean()) / production_missing_category.std()
zscore_r = zscore[abs(zscore) < 3]

# 对表production_missing_category进行归一化处理
new_production_missing_category = zscore_r / (zscore_r.max() - zscore_r.min())
print('归一化后的new_production_missing_category:')
new_production_missing_category['type'] = production_missing_category['type']
new_production_missing_category['fault_flg'] = production_missing_category['fault_flg']
print(new_production_missing_category)
```

```python
# 对表production_missing_num使用Z-Score方法
zscore = (production_missing_num - production_missing_num.mean()) / production_missing_num.std()
zscore_r = zscore[abs(zscore) < 3]

# 对表production_missing_num进行归一化处理
new_production_missing_num = zscore_r / (zscore_r.max() - zscore_r.min())
print('归一化后的new_production_missing_num:')
new_production_missing_num['fault_flg'] = production_missing_num['fault_flg']
new_production_missing_num['type'] = production_missing_num['type']
print(new_production_missing_num)
```

```python
# 对表production_missing_num_4_redshift使用Z-Score方法
zscore = (production_missing_num_4_redshift - production_missing_num_4_redshift.mean()) / production_missing_num_4_redshift.std()
zscore_r = zscore[abs(zscore) < 3]

# 对表production_missing_num_4_redshift进行归一化处理
new_production_missing_num_4_redshift = zscore_r / (zscore_r.max() - zscore_r.min())
print('归一化后的new_production_missing_num_4_redshift:')
new_production_missing_num_4_redshift['fault_flg'] = production_missing_num_4_redshift['fault_flg']
new_production_missing_num_4_redshift['type'] = production_missing_num_4_redshift['type']
print(new_production_missing_num_4_redshift)
```

```python
# 表production需要数据的规整化
production = new_production.replace(True,1)
production = new_production.replace(False,0)
production = new_production.replace(['A','B','C','D','E'],[1,2,3,4,5])

# 表production_missing_category需要数据的规整化
production_missing_category = new_production_missing_category.replace(True,1)
production_missing_category = new_production_missing_category.replace(False,0)
production_missing_category = new_production_missing_category.replace(['A','B','C','D','E'],[1,2,3,4,5])

# 表production_missing_num需要数据的规整化
production_missing_num = new_production_missing_num.replace(True,1)
production_missing_num = new_production_missing_num.replace(False,0)
production_missing_num = new_production_missing_num.replace(['A','B','C','D','E'],[1,2,3,4,5])

# 表production_missing_num_4_redshift需要数据的规整化
production_missing_num_4_redshift = new_production_missing_num_4_redshift.replace(True,1)
production_missing_num_4_redshift = new_production_missing_num_4_redshift.replace(False,0)
production_missing_num_4_redshift = new_production_missing_num_4_redshift.replace(['A','B','C','D','E'],[1,2,3,4,5])

print('表production:')
print(production)
print(production.describe())
print('\n\n\n')

print('表production_missing_category:')
print(production_missing_category)
print(production_missing_category.describe())
print('\n\n\n')

print('表production_missing_num:')
print(production_missing_num)
print(production_missing_num.describe())
print('\n\n\n')

print('表production_missing_num_4_redshift:')
print(production_missing_num_4_redshift)
print(production_missing_num_4_redshift.describe())
print('\n\n\n')
```

```python
# 求这一行的聚合值,这里以中位数为聚合值,并用这些值填充None
production_median = production.median()
print(production_median,end='\n\n')
production = production.fillna(production_median)
print(production)
```
